```python
#單次AC
user_ac = {}

while True:
    log = input()
    if log == "END":
        break
    user, prob, res = log.split()
    if user not in user_ac:
        user_ac[user] = {}
    if res == "AC":
        if prob not in user_ac[user]:
            user_ac[user][prob] = 0
        
max_ac = 0
ans = ""

for user in user_ac:
    if len(user_ac[user]) > max_ac:
        max_ac = len(user_ac[user])
        ans = user
  
print(ans)
```
