```python
#AC
score_board = {}
while True:
    k = input()
    if k == "DATA_INPUT_END":
        break
    deck = k.split()
    name = deck[0]
    score = deck[1]
    score_board[name] = score
while True:
    a = input()
    if a == "END":
        break    
    if a in score_board:
        print(score_board[a])
    else:
        print("gg")
```
