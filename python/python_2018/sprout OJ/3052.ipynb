```python
#單次AC
n = int(input())
score = {}

for i in range(n):
    
    name, status = input().split()
    
    if name not in score:
        score[name] = 1
    else:
        if status == "AC":
            score[name] += 1
    
max_score = 0
ans = ""

for name in score:
    if score[name] > max_score:
        max_score = score[name]
        ans = name
  
print(ans)
```
