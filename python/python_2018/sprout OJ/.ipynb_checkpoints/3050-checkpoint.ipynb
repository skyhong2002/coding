```python
#AC

'''

以下程式碼必須加到OJ裡面

import sys
import codecs
sys.stdout = codecs.getwriter("utf-8")(sys.stdout.detach())
sys.stdin = codecs.getreader("utf-8")(sys.stdin.detach())

'''

add_dict = {}
add_order = []
while True:
    address = input()
    if address == "END":
        break

    if address in add_dict:
        add_dict[address] += 1
    else:
        add_dict[address] = 1
        add_order.append(address)
        
for add in add_order:
    print(add_dict[add])
```
