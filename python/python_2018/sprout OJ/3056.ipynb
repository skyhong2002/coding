```python
#很難的AC

a = []
ans = []
cnt = int(input())
string = input()

for i in range(len(string)):
    a.append(string[i])
    
for i in range(cnt):
    for j in range(len(a)):
        if j > (len(a) - i - 1) // cnt:
            break
        ans.append(a[i % cnt + j * cnt])

for i in range(len(string)):
    print(ans[i], end = "")

print("") #換行
```
