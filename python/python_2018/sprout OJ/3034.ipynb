```python
#AC
list1 = list(map(int, input().split()))
list2 = list(map(int, input().split()))
Acnt = 0
Bcnt = 0

for i in range(len(list1)):
    for j in range(len(list2)):
        if list1[i] == list2[j]:
            if i == j:
                Acnt += 1
            else:
                Bcnt += 1
            break
print(str(Acnt)+"A"+str(Bcnt)+"B")
```
