```python
#AC in once

students = list(map(int, input().split()))
cnt = int(input())
check = 0

for i in range(cnt):
    qsn = int(input())
    for j in range(len(students)):
        if students[j] == qsn:
            print('YES')
            check = 1
            break
    if check == 0:
        print('NO')
    check = 0
```
