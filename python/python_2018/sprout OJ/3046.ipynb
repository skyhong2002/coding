```python
#單次AC

name = input()
article = input()
cnt = 0

name = name.lower()

a = article.lower() #全部變小寫
b = a.replace(",", " ")
c = b.replace(".", " ")
newarticle = c.split( )

for i in range(len(newarticle)):
    if newarticle[i] == name:
        cnt += 1

print(cnt)
```
