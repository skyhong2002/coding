```python
from PIL import Image, ImageFilter
```

```python
def rotate90(im):
    return img.rotate(90, expand=True)
```

```python
img = Image.open('sample.jpg')
img
```

```python
rotate90(img)
```

```python
def sobel_edge_detect(im):
    img = im.convert('L') #轉成灰階

    
    xkernel = (-1, 0, 1, -2, 0, 2, -1, 0, 1)
    x_filter = ImageFilter.Kernel((3, 3), xkernel, 4)
    im2 = img.filter(x_filter)
    im2.show()
    
    ykernel = (1, 2, 1, 0, 0, 0, -1, -2, -1)
    y_filter = ImageFilter.Kernel((3, 3), ykernel, 4)
    im3 = img.filter(y_filter)
    im3.show()
    '''
    W,H = im.size
    n_img = (W, H)
    for w in range(W):
        for h in range(H):
            sum_value = im2.getpixel((w, h)) + im3.getpixel((w, h))#將 im2 和 im3 的 pixel value 加起來
            if sum_value > 8 :
                n_img.putpixel((w, h), 255) #設成白色
            else:
                n_img.putpixel((w, h), 0) #設成黑色
    '''
    return 0 #n_img
```

```python
img2 = Image.open('sample2.jpg')
img2
```

```python
sobel_edge_detect(img2)
```
